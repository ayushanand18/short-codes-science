# Lissajous Curves

```python
import matplotlib.pyplot as p
%matplotlib inline
import math as m
import numpy as np
def lisfig(w1,w2,delta):
    pi=m.pi
    t=np.linspace(0,20,1000)
    x=np.cos(w1*t)
    y=np.cos(w2*t+(delta))
    p.plot(x,y)
    p.show()
```

```python
lisfig(1,2,pi/6)
```
